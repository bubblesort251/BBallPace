# Converting probability of winning the game into a sharpe ratio

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
r = 1
```

```python
data = pd.DataFrame([.01*i for i in range(1,101)], columns=['Ps'])
```

```python
data['ExpectedReturn'] = r*data['Ps'] - 1*(1-data['Ps'])
data['Variance'] = data['Ps']*(r - data['ExpectedReturn'])**2 + (1-data['Ps'])*(-1 - data['Ps'])**2
data['Vol'] = data['Variance']**(.5)
data['Sharpe'] = data['ExpectedReturn']/data['Vol']
```

```python
data.head(10)
```

```python
plt.plot(data['Ps'], data['Sharpe'], color='black')
plt.plot(data['Ps'], data['ExpectedReturn'], color='red')
plt.title('Sharpe Ratio and Expected Return verses win Prob, for r = .9')
plt.xlabel('Probability to Win Bet')
plt.legend(['Sharpe Ratio', 'ExpectedReturn'])
```
